***Step2: Feature Relationships***

```python
# Relation between gdpPerCapita and suicideRate
```

```python
sns.scatterplot(x='gdpPerCapita', y='suicideRate',hue='happinessScore', data=data)
```

***Note:*** Suicide Rate is high with low GDP/Capita and low Happiness Score

```python
#lets compare all the variables
sns.pairplot(data)
plt.show()
```

***The above pairplot shows that most of the variables have significant correlation***

```python
#Calculating the correlation cofficients
corr=data.corr()
corr
```

**We can see most of the correlation cofficients are close to 1 and -1.Those variables have good correlation**

```python
#lets test this with heat map
sns.heatmap(corr,annot=True)
```

**This also interpret the same results**

**Step 3: lets find the answers to some questions:**
- **Ques1:-** What are the top 10 countries having high suicide rate

```python
country_suicide_df=data[['Country','suicideRate']].sort_values(by='suicideRate',ascending=False).head(10)
sns.barplot(data=country_suicide_df,x='suicideRate',y='Country')
plt.show()
```

- **Ques2:-** Compare Life Expectancy of Male and Female

```python
life_men_women_df=data[['lifeExpectancyMen','lifeExpectancyWomen']].copy()
life_men_women_df.plot.area()
```
